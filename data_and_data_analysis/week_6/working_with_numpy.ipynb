#### Vectors

**Vector Addition**
- When you add vectors, you add each pair of elements.
- Bag A = [2, 3]
- Bag B = [1, 5]
- Total Items = [2+1, 3+5] = [3, 8]

```python
a = np.array([2, 3])
b = np.array([1, 5])
print("a + b =", a + b)
```

```python
# Vector addition
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

vector_sum = a + b
print(f"Vector a: {a}")
print(f"Vector b: {b}")
print(f"a + b: {vector_sum}")

# Geometric interpretation in 2D
a_2d = np.array([2, 3])
b_2d = np.array([1, 2])
sum_2d = a_2d + b_2d

plt.figure(figsize=(8, 6))
plt.quiver(0, 0, a_2d[0], a_2d[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005)
plt.quiver(0, 0, b_2d[0], b_2d[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.005)
plt.quiver(0, 0, sum_2d[0], sum_2d[1], angles='xy', scale_units='xy', scale=1, color='green', width=0.005)
plt.quiver(a_2d[0], a_2d[1], b_2d[0], b_2d[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.003, alpha=0.5)
plt.xlim(-1, 4)
plt.ylim(-1, 6)
plt.grid(True)
plt.title("Vector Addition: Parallelogram Rule")
plt.legend(['Vector a', 'Vector b', 'a + b'])
plt.show()
```

```python
# Properties of vector addition
c = np.array([1, -1, 2])

# Cummutative property: a + b = b + a
print(f"a + b = {a + b}")
print(f"b + a = {b + a}")
print(f"Cummutative: {np.array_equal(a + b, b + a)}")

# Associative property: (a + b) + c = a + (b + c)
left = (a + b) + c
right = a + (b + c)
print(f"(a + b) + c = {left}")
print(f"a + (b + c) = {right}")
print(f"Associative: {np.array_equal(left, right)}")
```

**Scalar Multiplication**
- Multiply every value in a vector by the same number.
- Think of multiplying prices by quantity.

```python
prices = np.array([100, 200])
discount = 0.5
print("Half price:", discount * prices)
```

```python
# Scalar multiplication
vector = np.array([2, 3, 1])
scalar = 3

scaled_vector = scalar * vector
print(f"Original vector: {vector}")
print(f"Scalar: {scalar}")
print(f"Scaled vector: {scaled_vector}")

# Geometric interpretation
original_2d = np.array([2, 1])
scales = [0.5, 1, 2, -1]

plt.figure(figsize=(10, 6))
colors = ['red', 'blue', 'green', 'purple']
for i, scale in enumerate(scales):
    scaled = scale * original_2d
    plt.quiver(0, 0, scaled[0], scaled[1], angles='xy', scale_units='xy', scale=1, color=colors[i], width=0.005, label=f"{scale} * v")

plt.xlim(-3, 5)
plt.ylim(-3, 3)
plt.grid(True)
plt.title("Scalar Multiplication effects")
plt.legend()
plt.show()
```

```python
# Properties of scalar multiplication
a = np.array([1, 2, 3])
c1, c2 = 2, 3

# associative: c1(c2*a) = (c1*c2)*a
left = c1 * (c2 * a)
right = (c1 * c2) * a
print(f"c1(c2*a) = {left}")
print(f"(c1*c2)*a = {right}")
print(f"Associative: {np.array_equal(left, right)}")

# Distributive over vector addition: c(a + b) = ca + cb
left = c1 * (a + b)
right = c1 * a + c1 * b
print(f"c(a + b) = {left}")
print(f"ca + cb = {right}")
print(f"Distributive: {np.array_equal(left, right)}")
```

**The Physicist's Perspective of a vector**
- Magnitude and Direction Vectors have both magnitude (length) and direction in space.

```python
import matplotlib.pyplot as plt
# Physicst's view: vectors as arrows in space
vector_a = np.array([3, 4])
magnitude = np.linalg.norm(vector_a)
print(f"Vector: {vector_a}")
print(f"magnitude: {magnitude}")

# Direction (unit vector)
unit_vector = vector_a / magnitude
print(f"Unit vector (direction): {unit_vector}")

# Visualization
plt.figure(figsize=(8, 6))
plt.quiver(0, 0, vector_a[0], vector_a[1], angles='xy', scale_units='xy', scale=1, color='red', width=0.005)
plt.quiver(0, 0, unit_vector[0], unit_vector[1], angles='xy', scale_units='xy', scale=1, color='blue', width=0.003)
plt.xlim(-1, 5)
plt.ylim(-1, 5)
plt.grid(True)
plt.title("Vector: Magnitude and Direction")
plt.legend(['Original Vector', 'Unit vector'])
plt.show()
```
